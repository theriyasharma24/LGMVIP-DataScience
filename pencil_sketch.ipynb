**Task 2**: Image to Pencil Sketch with Python

**Task Description** : In this task, an image is converted to a pencil sketch using cv2 library in python. The flow of the code is as follows:

**Original Image -> GrayScale Image -> Inverted GrayScale Image -> Blurry Image -> Final Image(Pencil Sketch)**

Importing modules cv2 and cv2_show.

```python
from google.colab import files
import cv2
from google.colab.patches import cv2_imshow
```

Displaying the Original Image.

```python
img=cv2.imread("mickeymouse.jpg")
cv2_imshow(img)
```

Converting Original Image into GrayScale Image .i.e. classic black and white image and then displaying it.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

Converting GrayScale Image into Inverted GrayScale Image or Negative Image and then displaying it.

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

Converting Inverted GrayScale Image to Inverted Blurry Image.

```python
img_smooth = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smooth)
```

After Dividing GrayScale Image by the Inverted Blurry Image, Final Image (pencil sketch) is obtained.

```python
img_final = cv2.divide(img_gray, 255 - img_smooth, scale=255)
cv2_imshow(img_final)
```

This is the final image which is the pencil sketch of original image.
